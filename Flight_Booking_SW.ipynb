```python
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
le = LabelEncoder()

# Iterate over the columns in the DataFrame
for col in df.columns:

    # If the column is categorical, encode it
    if df[col].dtype == "object":
        df[col] = le.fit_transform(df[col])

# Print the information about the DataFrame
print(df.info())
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,cmap="coolwarm")
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a list of variables
col_list = []
for col in df.columns:
    if ((df[col].dtype != 'object') & (col != 'price')):
        col_list.append(col)

# Calculate the VIF for each variable
X= df[col_list]
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                   for i in range(len(X.columns))]

# Print the DataFrame
print(vif_data)
```

```python
x = df.drop(columns=["price"])
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=14)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
difference=pd.DataFrame(np.c_[y_test,y_pred],columns=["Actual_Value","Predicted_Value"])
difference
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn import metrics
means_abs_error = metrics.mean_absolute_error(y_test,y_pred)
means_abs_error
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test, y_pred)
mean_sq_error=metrics.mean_squared_error(y_test,y_pred)
mean_sq_error
```

```python
root_mean_sq_error = np.sqrt(metrics.mean_squared_error(y_test,y_pred))
root_mean_sq_error
```

```python
sns.distplot(y_test,label="Actual")
sns.distplot(y_pred,label="Predicted")
plt.legend()
```
